# **Classifica√ß√£o com SVM (Margem R√≠gida e Suave)**

Os primeiros comandos abaixo importam ferramentas √∫teis para gerar conjuntos de dados sint√©ticos e para criar visualiza√ß√µes. Aqui est√° o que cada linha faz:

1. **`from sklearn.datasets import make_blobs`**:
   - Importa a fun√ß√£o `make_blobs` do m√≥dulo `sklearn.datasets`. 
   - Essa fun√ß√£o √© usada para criar dados sint√©ticos em formato de clusters (agrupamentos). √â muito √∫til para testar algoritmos de aprendizado de m√°quina ou visualiza√ß√µes.

2. **`import matplotlib.pyplot as plt`**:
   - Importa o m√≥dulo `pyplot` da biblioteca `matplotlib`, que √© usado para criar gr√°ficos e visualiza√ß√µes. 
   - Voc√™ usar√° o objeto `plt` para, por exemplo, criar gr√°ficos de dispers√£o, histogramas ou outros tipos de representa√ß√µes visuais dos dados.


````python
# Exemplo introdut√≥rio com make_blobs para visualiza√ß√£o de classifica√ß√£o bin√°ria
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt

# Gerando um conjunto de dados com duas classes
# X cont√©m os pontos e Y cont√©m os r√≥tulos (0 ou 1)
X, Y = make_blobs(n_samples=500, centers=2, random_state=0, cluster_std=0.5)

# Visualizando os dados em um gr√°fico de dispers√£o
plt.figure(figsize=(6, 5))
plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='spring', edgecolors='k')
plt.title('Problema de Classifica√ß√£o Bin√°ria com make_blobs')
plt.xlabel('Atributo 1')
plt.ylabel('Atributo 2')
plt.grid(True)
plt.show()

````

O que o SVM faz √© n√£o apenas tra√ßar uma linha entre as duas classes, mas tamb√©m considerar uma regi√£o ao redor dessa linha com uma certa largura.
O c√≥digo a seguir ilustra isso:

````python
# Criando uma linha entre -1 e 3.5 para o eixo x
xfit = np.linspace(-1, 3.5)

# Reexibindo os dados originais
plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='spring')

# Tra√ßando retas com diferentes inclina√ß√µes, interceptos e margens
for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')  # linha preta
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none', 
                     color='#AAAAAA', alpha=0.4)  # faixa cinza representando a margem

plt.xlim(-1, 3.5)
plt.title("Exemplos de retas separadoras com margem")
plt.xlabel('Atributo 1')
plt.ylabel('Atributo 2')
plt.grid(True)
plt.show()
````

# **C√≥digos dos Classificadores**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# N√∫mero de amostras desejado
num_samples = 569  # Defina o n√∫mero de amostras que deseja utilizar

# Lista de pares de features para teste
feature_sets = [
    ('mean perimeter', 'mean concave points'),
    ('mean radius', 'mean texture'),
    ('mean symmetry', 'mean fractal dimension')
]

# Selecionar uma amostra do dataset
df_sampled = df.sample(n=num_samples, random_state=42)  # Seleciona `num_samples` amostras aleat√≥rias

# Loop pelos pares
for f1, f2 in feature_sets:
    print(f"\n Avalia√ß√£o para: {f1} + {f2}")
    
    # Dados e r√≥tulo
    X = df_sampled[[f1, f2]]
    y = df_sampled['target']

    # Dividir em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Modelo SVM linear
    model = SVC(kernel='linear')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # M√©tricas
    acc = accuracy_score(y_test, y_pred)
    print(f"Acur√°cia: {acc:.4f}")
    print(classification_report(y_test, y_pred, target_names=["Maligno", "Benigno"], zero_division=0))

    # # Matriz de Confus√£o
    # cm = confusion_matrix(y_test, y_pred)
    # sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Maligno", "Benigno"], yticklabels=["Maligno", "Benigno"])
    # plt.title(f"Matriz de Confus√£o - {f1} vs {f2}")
    # plt.xlabel("Predito")
    # plt.ylabel("Real")
    # plt.show()

    # Fronteira de decis√£o
    plt.figure(figsize=(8, 6))
    plt.scatter(X_test[f1], X_test[f2], c=y_test, cmap='coolwarm', s=50, edgecolors='k', alpha=0.7)

    # Gera√ß√£o da grade
    x_min, x_max = X_test[f1].min() - 0.01, X_test[f1].max() + 0.01
    y_min, y_max = X_test[f2].min() - 0.01, X_test[f2].max() + 0.01
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    
    # Converter a grade para DataFrame com os mesmos nomes de colunas
    grid = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=[f1, f2])
    Z = model.predict(grid)
    Z = Z.reshape(xx.shape)

    # Plot da fronteira
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.xlabel(f1)
    plt.ylabel(f2)
    plt.title(f"Fronteira de Decis√£o - {f1} vs {f2}")
    plt.colorbar(label='0 = Maligno, 1 = Benigno')
    plt.show()
````

üìå **Coluna `support`**

A coluna **`support`** representa o **n√∫mero de amostras reais (verdadeiras) da classe no conjunto de teste**.

No exemplo acima:

- **Maligno:** 65 amostras verdadeiras
- **Benigno:** 106 amostras verdadeiras
- **Total:** 171 amostras no conjunto de teste

> üí° Serve como base para calcular m√©dias ponderadas e tamb√©m para entender a distribui√ß√£o das classes (balanceamento).


üìå **Linha `macro avg`**

A linha **`macro avg`** calcula a **m√©dia simples** (n√£o ponderada) das m√©tricas (precision, recall, f1-score) entre todas as classes.

- **N√£o leva em conta o n√∫mero de exemplos por classe**
- √ötil quando queremos **dar igual import√¢ncia a cada classe**, mesmo em datasets **desequilibrados**

Para o exemplo mean perimeter + mean concave points:

- `macro avg` precision = m√©dia de (0.94, 0.86) = 0.90 
- `macro avg` recall = m√©dia de (0.74, 0.97) = 0.86
- `macro avg` f1-score = m√©dia de (0.83, 0.91) = 0.87



üìå **Linha `weighted avg`**

A linha **`weighted avg`** calcula a m√©dia **ponderada pelo n√∫mero de amostras de cada classe** (coluna `support`).

> **Usa a f√≥rmula:**
>
> $$
 \text{weighted avg} = \frac{\sum_i (\text{support}_i \times \text{m√©trica}_i)}{\sum_i \text{support}_i}
  $$

Exemplo para `precision`:

$$
\text{precision}_\text{weighted} = \frac{65 \cdot 0.94 + 106 \cdot 0.86}{65 + 106} ‚âà 0.89
$$

Essa m√©dia reflete o desempenho **global real**, considerando a propor√ß√£o de cada classe.


‚úÖ **Resumo**

| Linha/Coluna       | Significado                                                                 |
|--------------------|-----------------------------------------------------------------------------|
| `support`          | N√∫mero de amostras verdadeiras por classe no conjunto de teste              |
| `macro avg`        | M√©dia simples das m√©tricas entre as classes                                 |
| `weighted avg`     | M√©dia ponderada das m√©tricas, levando em conta o n√∫mero de amostras         |


# **C√≥digo para o modelo com todas as 30 features**

````python
from sklearn.preprocessing import StandardScaler

# Separando features (todas menos 'target') e r√≥tulo
X_all = df.drop(columns=['target'])
y_all = df['target']

# Padronizar os dados (importante para SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_all)

# Dividir em treino/teste
X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(X_scaled, y_all, test_size=0.3, random_state=42)

# SVM Linear
model_all = SVC(kernel='linear')
model_all.fit(X_train_all, y_train_all)
y_pred_all = model_all.predict(X_test_all)

# Avalia√ß√£o
print("Modelo com TODAS as 30 FEATURES")
acc_all = accuracy_score(y_test_all, y_pred_all)
print(f"Acur√°cia: {acc_all:.4f}")
print("Relat√≥rio de Classifica√ß√£o:")
print(classification_report(y_test_all, y_pred_all, target_names=["Maligno", "Benigno"]))

# # Matriz de confus√£o
# cm_all = confusion_matrix(y_test_all, y_pred_all)
# sns.heatmap(cm_all, annot=True, fmt='d', cmap='Blues', xticklabels=["Maligno", "Benigno"], yticklabels=["Maligno", "Benigno"])
# plt.title("Matriz de Confus√£o - TODAS AS FEATURES")
# plt.xlabel("Predito")
# plt.ylabel("Real")
# plt.show()
````

‚öñÔ∏è **Trade-off entre n√∫mero de features e complexidade**

| Fator                           | Com poucas features                          | Com muitas features (30 ou mais)                  |
|--------------------------------|----------------------------------------------|--------------------------------------------------|
| üß† **Capacidade de generaliza√ß√£o** | Pode ser limitada (pouca informa√ß√£o)         | Potencialmente maior, se n√£o houver overfitting |
| üöÄ **Tempo de treino/teste**       | R√°pido                                       | Pode crescer linear ou at√© exponencialmente     |
| üßÆ **Complexidade computacional**  | Baixa (menor n√∫mero de opera√ß√µes)            | Alta (mais vari√°veis, mais mem√≥ria)             |
| üìâ **Risco de overfitting**       | Menor (menos vari√°veis para memorizar ru√≠do) | Maior (sobretudo em conjuntos de dados pequenos) |
| üîç **Interpretabilidade**          | Alta (f√°cil de entender e visualizar)         | Baixa (modelo vira uma ‚Äúcaixa preta‚Äù)           |


üß† **Ent√£o, como selecionar features?**

Usar todas as features **pode ser ben√©fico**, como vimos no seu caso, mas **nem sempre √© o ideal**. Para equilibrar desempenho e simplicidade, usamos **sele√ß√£o de features**, com abordagens como:

#### 1. **M√©todos Estat√≠sticos**
- `SelectKBest`, `SelectPercentile` com testes como ANOVA (`f_classif`), `chi2`.
- Exemplo: selecionar as 10 melhores features com maior correla√ß√£o com o r√≥tulo.

#### 2. **M√©todos baseados em modelo**
- Random Forest ou Linear SVM com `coef_` ou `feature_importances_` para ranquear.
- `Recursive Feature Elimination (RFE)` para eliminar as menos importantes iterativamente.

#### 3. **M√©todos de Embedded Learning**
- L1 Regularization (`Lasso`, `LinearSVC` com `penalty='l1'`) que zera coeficientes irrelevantes.


# **Par√¢metro C no SVM**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Sele√ß√£o das duas melhores features
features = ['mean perimeter', 'mean concave points']
X = df[features]
y = df['target']

# Normaliza√ß√£o
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Divis√£o treino/teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Lista de valores de C
C_values = [100, 10, 1, 0.1, 0.01,0.001]
n = len(C_values)
ncols = 3
nrows = (n + 1) // ncols

# Criar subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 4 * nrows))
axes = axes.flatten()

# Tabela de resultados
tabela_resultados = []
model = SVC(kernel='linear')
for idx, C in enumerate(C_values):
    model.set_params(C=C)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # M√©tricas
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)

    tabela_resultados.append({
        "C": C,
        "Accuracy": round(acc, 4),
        "Precision (macro avg)": round(report['macro avg']['precision'], 4),
        "Recall (macro avg)": round(report['macro avg']['recall'], 4),
        "F1-score (macro avg)": round(report['macro avg']['f1-score'], 4)
    })

    # Fronteira de decis√£o e margens
    ax = axes[idx]
    x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
    y_min, y_max = X_scaled[:, 1].min() - 0.5, X_scaled[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    ax.contourf(xx, yy, Z, cmap='coolwarm', alpha=0.3)
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='coolwarm', edgecolors='k', s=40)

    # Plot da fronteira de decis√£o e margens (reta linear)
    w = model.coef_[0]
    b = model.intercept_[0]
    slope = -w[0] / w[1]
    intercept = -b / w[1]
    margin = 1 / np.linalg.norm(w)

    x_vals = np.linspace(x_min, x_max, 200)
    y_vals = slope * x_vals + intercept
    y_margin_up = y_vals + margin
    y_margin_down = y_vals - margin

    ax.plot(x_vals, y_vals, 'k-', label='Fronteira')
    ax.plot(x_vals, y_margin_up, 'k--', linewidth=1, label='Margem')
    ax.plot(x_vals, y_margin_down, 'k--', linewidth=1)

    ax.set_title(f'C = {C}')
    ax.set_xlabel(features[0])
    ax.set_ylabel(features[1])
    ax.legend(loc='upper right')

# Remover subplots n√£o usados (se C_values < ncols*nrows)
for i in range(len(C_values), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# Mostrar tabela final com resultados
df_resultados = pd.DataFrame(tabela_resultados)
display(df_resultados)
````

Com base na tabela fornecida, que apresenta os resultados da acur√°cia, precis√£o, recall e F1-score (m√©dias macro) para diferentes valores do par√¢metro `C` no SVM linear usando duas features (`mean perimeter` e `mean concave points`), podemos concluir o seguinte:


 üìä **An√°lise dos Resultados**

| C        | Accuracy | F1-score (macro avg) | Observa√ß√£o                                   |
|----------|----------|----------------------|----------------------------------------------|
| **100**  | 0.9064   | 0.8995               | Margem r√≠gida; desempenho est√°vel            |
| **10**   | 0.9064   | 0.8995               | Igual ao anterior; margem ainda r√≠gida       |
| **1**    | 0.9064   | 0.8995               | Idem; modelo pouco sens√≠vel a varia√ß√µes aqui |
| **0.1**  | 0.9181   | 0.9114               | Melhor desempenho; margem suave efetiva      |
| **0.01** | 0.9181   | 0.9087               | Leve queda no F1; ainda bom                  |
| **0.001**| 0.7018   | 0.5645               | Subajuste (underfitting); margem larga demais|



 ‚úÖ **Conclus√µes**

1. **Valores altos de `C` (‚â• 1)** resultam em um **modelo com margem r√≠gida**, ou seja, o classificador tenta separar os dados da forma mais exata poss√≠vel, evitando erros. O desempenho √© est√°vel, mas n√£o √≥timo ‚Äî o modelo pode estar ‚Äúfor√ßando‚Äù a separa√ß√£o com margens muito estreitas.

2. **Valores intermedi√°rios de `C` (0.1 e 0.01)** permitem **margens mais amplas**, aceitando alguns erros para obter uma **separa√ß√£o mais est√°vel e generaliz√°vel**. Esses valores apresentaram os **melhores resultados de acur√°cia e F1-score**, o que indica que o modelo teve **melhor capacidade de generaliza√ß√£o** ao tolerar um pequeno grau de erro.

3. **Valor muito pequeno de `C` (0.001)** torna o modelo **excessivamente permissivo**. Ele foca tanto em ampliar a margem que acaba **ignorando o sinal das classes**, resultando em **baixo recall, acur√°cia e F1-score**. Isso caracteriza **underfitting** ‚Äî o modelo √© simples demais para capturar a separa√ß√£o real entre as classes.


üí¨ **Resumo:**

> *"Essa tabela mostra como o par√¢metro `C` no SVM afeta o equil√≠brio entre margem e erro. Quando `C` √© grande, o modelo tenta n√£o errar, mas pode memorizar ru√≠do. Quando `C` √© pequeno, ele foca em generalizar bem, mas pode errar demais se exagerar. O melhor desempenho foi obtido com `C = 0.1`, mostrando que permitir algum erro pode ser vantajoso quando temos dados sobrepostos."*
